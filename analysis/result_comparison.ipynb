```python
import pandas as pd
import numpy as np
```

```python
old_coef = pd.read_csv('/data/tide/projects/ho_infxn_ml/results/model_results/20241105/clr_coefficients.csv')
old_coef
```

```python
new_coef = pd.read_csv('/data/tide/projects/ho_infxn_ml/results/model_results/20241207/clr_coefficients_removed_prior_pathogen.csv')
new_coef.columns = ['match','target','variable','ppr_coef','ppr_lower_CI','ppr_upper_CI','ppr_SE_coef','ppr_pval','ppr_sig_flag']
new_coef
```

```python
combined_coef = old_coef.merge(new_coef,
                               on = ['match','target','variable'],
                               how = 'right')
combined_coef
```

```python
combined_coef.to_csv('/data/tide/projects/ho_infxn_ml/results/model_results/20241207/clr_coefficients_comparison.csv',
                    index = False)
```
